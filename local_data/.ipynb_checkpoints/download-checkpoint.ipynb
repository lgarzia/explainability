```python
import datetime 
import dbm
import uuid
from pathlib import Path
import pandas as pd
from bs4 import BeautifulSoup
from splinter import Browser
from selenium.webdriver.chrome.service import Service
import time
import navigate
import extract
```

```python
my_service = Service(executable_path='./chromedriver.exe')
browser = Browser('chrome', service=my_service)
```

```python
url = 'https://www.mshp.dps.missouri.gov/HP68/SearchAction'
browser.visit(url)
```

```python
available_dates = navigate.get_available_date_list(browser)
```

```python
data_path = Path('./data')
data_path.mkdir(parents=True, exist_ok=True)
```

```python
date_check_dbm = dbm.open(str(data_path / "date_check_dbm"), 'c')
```

```python
for date in available_dates:
    if date_check_dbm.get(date[-1]):
        print(f"skipping - already completed - {date[-1]}")
    else:
        print(f"Starting - {date[-1]}")
        date_data_path = data_path / f"{date[-1]}"
        date_data_path.mkdir(parents=True, exist_ok=True)
        navigate.enter_available_date(browser, date[0])
        time.sleep(3)
        navigate.search_crash_report(browser)
        time.sleep(5)
        df_all, anchors = extract.build_and_save_daily_crash_report(browser, date_data_path)
        for i, anchor in enumerate(anchors):
            record_uuid = df_all.iloc[i]["acc_uuid"] # 
            #print(df_all.iloc[i]["raw_href"], df_all.iloc[i]["Name"])
            time.sleep(1)
            record_detail_url = df_all.iloc[i]["raw_href"]
            browser.visit(record_detail_url)
            results = extract.build_and_save_daily_crash_report_details(browser, date_data_path, record_uuid)
        date_check_dbm[date[-1]] = 'Y'
        print(f"Completed - {date[-1]}")
        time.sleep(10)
    browser.visit(url)
```
